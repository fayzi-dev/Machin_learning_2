```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data 
# y = 2*x -1
x = np.array([1,2,3,4])
y = np.array([1,3,5,7])
```

```python
x.shape
```

```python
x = np.array([1,2,3,4]).reshape(-1,1)
```

```python
x.shape
```

```python
#plot
plt.plot(x, y, '*')
plt.show()
```

```python
#model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
# y = m*x + c
m = model.coef_[0]
c = model.intercept_
print(m,c)
```

```python
#plot
plt.plot(x, y, '*')
plt.plot(x, model.predict(x))
plt.show()
```

```python
result = model.predict([[5]])
result
```

```python
#Example 2 
years_of_experience = np.array([0, 1, 2, 4, 5, 8, 9, 10]).reshape(-1,1)
salaries = np.array([15000, 25000, 35000, 55000, 70000, 85000, 105000, 120000])
```

```python
years_of_experience.shape
```

```python
plt.plot(years_of_experience, salaries, '*')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(years_of_experience,salaries)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
y_pred = model.predict(years_of_experience)
y_pred
```

```python
plt.plot(years_of_experience, salaries, '*')
plt.plot(years_of_experience, y_pred)
plt.show()
```

```python
print(model.score(years_of_experience, salaries) *100 )
```
