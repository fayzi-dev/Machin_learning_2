```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,4,3,7,8,1,3,9,11]
plt.plot(x, y, color='red')
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(x, y,'o', color='blue')
plt.show()
```

```python
plt.plot(x, y,marker='*', color='blue')
plt.show()
```

```python
Name = ['Mohammad', 'Ali', 'Hassan', 'Ahmad']
Salary = [1000, 2000, 1500, 800]

plt.bar(Name,Salary, color='black')
plt.title('Salary')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```

```python
Name = ['Mohammad', 'Ali', 'Hassan', 'Ahmad']
Salary = [1000, 2000, 1500, 800]

plt.pie(Salary, labels=Name, autopct='%1.1f%%', colors=['red','blue','green','yellow'])
plt.show()
```

```python
import numpy as np
x = np.random.rand(1000,1)
plt.hist(x, bins=5)
plt.show()
```

```python
X = np.random.rand(64, 64)
plt.imshow(X, cmap='hot')
plt.colorbar()
plt.show()
```
